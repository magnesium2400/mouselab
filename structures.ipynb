# See information about structures and structure sets

This notebook can be used to visualise information about structures/structure sets.

## Prelims

```python
from pathlib import Path
from allensdk.core.mouse_connectivity_cache import MouseConnectivityCache
from allensdk.api.queries.ontologies_api import OntologiesApi
import pandas as pd
import json
```

## See all structure sets

```python
mcc = MouseConnectivityCache(manifest_file='./manifest.json')
structure_tree = mcc.get_structure_tree()
oapi = OntologiesApi()

# get the ids of all the structure sets in the tree
structure_set_ids = structure_tree.get_structure_sets()

# query the API for information on those structure sets
pd.DataFrame(oapi.get_structure_sets(structure_set_ids))
```

### Save to json

```python
with open(f'structures/structure_sets.json', 'w') as f: 
    json.dump(oapi.get_structure_sets(structure_set_ids), f)
```

## See all structures within a structure set

```python
# look at structures/structure_sets.csv to see all structure set id's you can use
structure_set_id = 688152357
summary_structures = structure_tree.get_structures_by_set_id([structure_set_id])
pd.DataFrame(summary_structures)
```

### Save to json

```python
with open(f'structures/{structure_set_id}.json', 'w') as f: 
    json.dump(summary_structures, f)
```
